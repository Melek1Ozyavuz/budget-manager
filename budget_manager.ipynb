```python
# Follow the given total budget
budget=20000

# Get inputs from user for expenses
groceries = float(input("How much did you spend on groceries?"))
entertainment = float(input("How much did you spend on entertainment?"))
transport = float(input("How much did you spend on transport?"))
savings = float(input("How much did you save?"))

# Calculate total expenditure and remaining budget
total_expenditure = groceries + entertainment + transport + savings
remaining_budget = budget - total_expenditure

# Calculate the usage percentage based on the total budget
groceries_percentage = groceries/budget
entertainment_percentage = entertainment/budget
transport_percentage = transport/budget
savings_percentage = savings/budget

# Display the results
print(f"Total expenditure is:{total_expenditure:.2f}")
print(f"Remaining budget is:{remaining_budget:.2f}")
print(f"Percentage of groceries is:{groceries_percentage:.2%}")
print(f"Percentage of entertainment is:{entertainment_percentage:.2%}")
print(f"Percentage of transport is:{transport_percentage:.2%}")
print(f"Percentage of savings is:{savings_percentage:.2%}")

```
