```python
import pandas as pd

# PART 1: DATA LOADING AND INITIAL SETUP

try:
    df = pd.read_csv("data.csv")
except FileNotFoundError:
    print("Error: 'data.csv' file not found. Please make sure it is uploaded.")
    exit()



# PART 2: DATA CLEANING AND EXPLORATORY DATA ANALYSIS

print("\n- Data Overview -")
print(f"Shape (rows, columns): {df.shape}")
print("\nData types and missing value summary:")
df.info()

df["Name"] = df["Name"].astype(str).str.lower()  #standardize text
df_clean = df.dropna(subset=["Height"])  #handling missing datas



# PART 3: FILTERING

print("\n- Category Frequency Analysis -")
type_counts = df["Type1"].value_counts().head(5)
print("Top 5 most frequent Type1 categories:")
print(type_counts)


print("\n- Conditional Filtering -")
fireflyer_pokemon = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")]
print(fireflyer_pokemon)

print("\n- Groupby Stats -")
group=df.groupby("Type1")
print(group["Height"].mean())






```
